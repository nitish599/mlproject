```python
#dependencies

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import warnings
from core.yolov4 import filter_boxes
import core.utils as utils
import time
from tensorflow.python.saved_model import tag_constants
warnings.filterwarnings('ignore') 
```

```python
#loading the saved model
PATH_TO_SAVED_MODEL="C:/Users/nitis/tensorflow-yolov4-tflite-master/checkpoints/custom-608"
print('Loading model...', end='')
# Load saved model and build the detection function
loaded_model=tf.saved_model.load(PATH_TO_SAVED_MODEL, tags=[tag_constants.SERVING])
print("Done")
```

```python
infer = loaded_model.signatures["serving_default"]
```

```python
# #getting frames from webcam through opencv2
vid = cv2.VideoCapture(0) 
vid.set(cv2.CAP_PROP_FRAME_WIDTH, 608)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 608)
while(True): 
    # Capture the video frame 
    # by frame  
    start_time = time.time()
    ret, frame = vid.read() 
    resized = cv2.resize(frame, (608,608), interpolation = cv2.INTER_AREA)
    # Display the resulting frame 
    input_tensor=tf.convert_to_tensor(resized)
    input_tensor = tf.cast(input_tensor,dtype=tf.float32)
    #print(type(input_tensor))
    # The model expects a batch of images, so add an axis with `tf.newaxis`.
    input_tensor=input_tensor[tf.newaxis, ...]
    # input_tensor = np.expand_dims(image_np, 0) 
    detections=infer(input_tensor)
    for key,value in detections.items():
        boxes = value[:, :, 0:4]
        pred_conf = value[:, :, 4:]
        print(boxes)

    #print(resized_frame.shape)
        boxes, scores, classes, valid_detections = tf.image.combined_non_max_suppression(
            boxes=tf.reshape(boxes, (tf.shape(boxes)[0], -1, 1, 4)),
            scores=tf.reshape(
                pred_conf, (tf.shape(pred_conf)[0], -1, tf.shape(pred_conf)[-1])),
            max_output_size_per_class=50,
            max_total_size=50,
            iou_threshold=0.45,
            score_threshold=0.25
        )
    pred_bbox = [boxes.numpy(), scores.numpy(), classes.numpy(), valid_detections.numpy()]
#     print(pred_bbox)
    image = utils.draw_bbox(frame, pred_bbox)
    fps = 1.0 / (time.time() - start_time)
    print("FPS: %.2f" % fps)
    result = np.asarray(image)
    cv2.namedWindow("result", cv2.WINDOW_AUTOSIZE)
    cv2.imshow('result', result) 
    # the 'q' button is set as the 
    # quitting button you may use any 
    # desired button of your choice 
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
# After the loop release the cap object 

vid.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```
